```python
import re

def parse_stat(val):
    if isinstance(val, str):
        match = re.match(r"(\d+)([+-])(\d+)", val)
        if match:
            base, sign, offset = match.groups()
            if sign == "+":
                return int(base) + int(offset)
            elif sign == "-":
                return int(base) - int(offset)
        elif val.isdigit():
            return int(val)
    return None  # Trường hợp không khớp
```

```python
columns_to_fix = [
    "Overall", "Potential", "Crossing", "Finishing", "Heading accuracy",
    "Short passing", "Volleys", "Dribbling", "Curve", "Long passing",
    "Ball control", "Acceleration", "Sprint speed", "Reactions",
    "Shot power", "Jumping", "Stamina", "Strength", "Long shots",
    "Aggression", "Interceptions", "Att. Position", "Vision",
    "Defensive awareness", "Standing tackle", "GK Diving", "GK Handling",
    "GK Positioning", "GK Reflexes"
]

for col in columns_to_fix:
    df[col] = df[col].apply(parse_stat)
```

```python
def extract_cm(height_str):
    if isinstance(height_str, str):
        match = re.match(r"(\d+)\s*cm", height_str)
        if match:
            return int(match.group(1))
    return None

df["Height"] = df["Height"].apply(extract_cm)
```

```python
def extract_kg(weight_str):
    if isinstance(weight_str, str):
        match = re.match(r"(\d+)\s*kg", weight_str)
        if match:
            return int(match.group(1))
    return None

df["Weight"] = df["Weight"].apply(extract_kg)
```

```python
def parse_euro(value_str):
    if isinstance(value_str, str):
        value_str = value_str.replace('€', '').strip()
        if value_str.endswith('M'):
            return float(value_str[:-1]) * 1_000_000
        elif value_str.endswith('K'):
            return float(value_str[:-1]) * 1_000
        elif value_str.isdigit():
            return float(value_str)
    return 0.0  # Trường hợp '€0' hoặc không xác định

df['Value'] = df['Value'].apply(parse_euro)
df['Wage'] = df['Wage'].apply(parse_euro)
```

```python
import pandas as pd
import numpy as np

# Load dữ liệu
df = pd.read_csv("/content/DataFC25.csv",encoding='utf-8-sig')

# Hàm chuyển "81+2" => 83, hoặc giữ nguyên nếu chỉ là số
def convert_to_numeric(value):
    if isinstance(value, str):
        parts = value.split("+")
        try:
            numbers = [int(part.strip()) for part in parts]
            return sum(numbers)
        except ValueError:
            return np.nan
    try:
        return float(value)
    except ValueError:
        return np.nan

# Xác định các cột dạng object có khả năng chứa số
object_cols = df.select_dtypes(include='object').columns

# Tạo dict lưu tên cột được chuyển đổi
converted_columns = {}

# Áp dụng chuyển đổi cho các cột khả nghi
for col in object_cols:
    sample = df[col].dropna().astype(str).iloc[0]
    if any(char.isdigit() for char in sample):
        df[col + "_cleaned"] = df[col].apply(convert_to_numeric)
        converted_columns[col] = col + "_cleaned"

# Thay thế cột gốc bằng cột đã làm sạch
for original_col, cleaned_col in converted_columns.items():
    df[original_col] = df[cleaned_col]
    df.drop(columns=[cleaned_col], inplace=True)

# Lưu lại file sau khi xử lý nếu cần
df.to_csv("data_cleaned_processed.csv", index=False)
```
